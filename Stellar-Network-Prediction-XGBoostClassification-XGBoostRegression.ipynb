```python
stellar_df['transaction_base_price'] = stellar_df['fee_charged'] / stellar_df['transaction_operation_count']
```

```python
stellar_df['surge_price_ind'].value_counts()
```

```python
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler
from collections import Counter
```

```python
stellar_df.columns

features_ = [ 
        'prior_successful_transaction_count',
       'prior_failed_transaction_count', 
        'prior_operation_count',
       'prior_successful_operation_count', 
        'prior_max_fee_charged',
       'prior_min_fee_charged', 
        'prior_avg_fee_charged',
'transaction_operation_count',
    'max_fee_bid', 'new_max_fee_bid'
]

X_train = stellar_df[features_]

y_train = stellar_df['surge_price_ind']
```

```python
Counter(y_train)
```

```python
rus = RandomUnderSampler() 

X_train_rus, y_train_rus = rus.fit_resample(X_train, y_train)

print(Counter(y_train_rus))
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

numeric_pipeline = Pipeline(
    steps=[("impute", SimpleImputer(strategy="mean")), 
           ("scale", StandardScaler())]
)



```

```python
num_cols = X_train.select_dtypes(include="number").columns

from sklearn.compose import ColumnTransformer

full_processor = ColumnTransformer(
    transformers=[
        ("numeric", numeric_pipeline, num_cols),
    ]
)
```

```python
X_train_processed = full_processor.fit_transform(X_train)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_train_processed, y_train, stratify=y_train, random_state=1121218
)

```

```python
from sklearn.metrics import accuracy_score
import xgboost as xgb

# Init classifier
xgb_cl = xgb.XGBClassifier(tree_method='gpu_hist', gpu_id=0,verbosity=3,random_state=42)

# Fit
xgb_cl.fit(X_train, y_train)

# Predict
preds = xgb_cl.predict(X_test)

# Score
accuracy_score(y_test, preds)
```

### Regression for Non Surge

```python
features_ = [ 
        'prior_successful_transaction_count',
       'prior_failed_transaction_count', 
        'prior_operation_count',
       'prior_successful_operation_count', 
        'prior_max_fee_charged',
       'prior_min_fee_charged', 
        'prior_avg_fee_charged',
'transaction_operation_count',
    'max_fee_bid', 'new_max_fee_bid'
]

target = ['fee_charged']

X_reg_train = stellar_df[stellar_df['surge_price_ind'] == 1 ][features_]


y_reg_train = stellar_df[stellar_df['surge_price_ind'] == 1 ][target]
```

```python
y_reg_train = y_reg_train / 100 
```

```python
numeric_reg_pipeline = Pipeline(
    steps=[("impute", SimpleImputer(strategy="mean")), 
           ("scale", StandardScaler())]
)


num_cols = X_reg_train.select_dtypes(include="number").columns

full_reg_processor = ColumnTransformer(
    transformers=[
        ("numeric", numeric_pipeline, num_cols),
    ]
)

X_reg_train_processed = full_reg_processor.fit_transform(X_reg_train)


```

```python
from sklearn.metrics import mean_squared_error
import xgboost as xgb


# Init classifier
xgb_reg = xgb.XGBRegressor(tree_method='gpu_hist', 
                           gpu_id=0,
                           verbosity=3,
                           random_state=42,
                           n_estimators=128,
                           eval_metric = mean_squared_error)

# Fit
xgb_reg.fit(X_reg_train_processed, y_reg_train)


# Predict
reg_preds = xgb_reg.predict(X_reg_train_processed)

# Score
mean_squared_error(y_reg_train, reg_preds)
```
